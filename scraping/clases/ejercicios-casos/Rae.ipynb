```python
import requests
from bs4 import BeautifulSoup
import re

def get_acepciones(word):
    headers = {'Accept':'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
    'Accept-Encoding':'gzip, deflate, sdch',
    'Accept-Language':'es-ES,es;q=0.8,en-US;q=0.6,en;q=0.4',
    'Cache-Control':'max-age=0',
    'Connection':'keep-alive',
    'Content-Length':'189',
    'Content-Type':'application/x-www-form-urlencoded',
    'Host':'dle.rae.es',
    'Referer':'http://dle.rae.es/srv/search?w=a&m=form',
    'Upgrade-Insecure-Requests':'1',
    'Origin':'http://dle.rae.es',
    'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/57.0.2987.133 Safari/537.36'}
    acepciones = []
    ht = requests.get('http://dle.rae.es/srv/search?w={0}&m=form'.format(word))
    print(ht.text)
    bs = BeautifulSoup(ht.content, 'html.parser')
    print(ht.content)
    for redir in bs.find_all('a'): 
        fetch = re.findall(r'"(fetch\?id=[\s\S]+?)"', str(redir))[0]
        ac = requests.get('http://dle.rae.es/srv/' + fetch, headers = headers)
        acbs = BeautifulSoup(ac.content, 'html.parser')
        acepciones.append(acbs)
    #extract defs de bs
    if len(acepciones) == 0:
        acepciones.append(bs)
    for raw in acepciones:
        raw = raw.find_all('p')
    return acepciones

a = get_acepciones('hola')
print(a)
```
