## 1.2. Type Hinting


&nbsp;&nbsp;&nbsp;&nbsp; Vendido! Ahora, 驴c贸mo hago para que se entiendo que es lo que la funci贸n espera?. Para lograr eso vamos a utilizar **Type hints**. Si prestaste atenci贸n a la imagen anterior ya habr谩s podido ver uno. El formato es `':'` seguido del tipo que esperamos sea nuestra variable (en nuestro caso el `'bool'`, `'= False'` es el valor por defecto).

&nbsp;&nbsp;&nbsp;&nbsp; Los type hints fueron introducidos en el [PEP 484](https://peps.python.org/pep-0484/) y si bien las anotaciones no eran nada nuevo para Python, lo que estableci贸 fue el standard de c贸mo indicar tipos para que se puede hacer un an谩lisis est谩tico de tipos; asi como, por ejemplo, introducir la libreria standard **Typing** que nos ayuda a espeficar tipos.

&nbsp;&nbsp;&nbsp;&nbsp; Dicho esto, modifiquemos el c贸digo para que sea m谩s interpretable:


````python
def multi_call(num: int, alt: bool = False) -> int:
    if alt:
        return num * 5
    return num * - 2

b = multi_call(23)
````

&nbsp;&nbsp;&nbsp;&nbsp; Ok, me siento estafado. 驴Por qu茅 me acepta un float como par谩metro cuando le especifiqu茅 que ten铆a que ser entero? Como mencionamos anteriormente, lo que hacemos con type hint no es forzar validaci贸n de tipos sino hacer un anotado de modo que seamos nosotros (o una herramienta) los que controlemos con que llamamos a que y con que.

&nbsp;&nbsp;&nbsp;&nbsp; Si solo quisieramos aceptar enteros tendr铆amos que agregar el control a mano:

````python
def multi_call_control(num: int, alt: bool = False) -> int:
    assert isinstance(num, int), f"Invalid type :{type(num)}"
    if alt:
        return num * 5
    return num * - 2
print(multi_call_control(3.5))
````

Con esto me refiero a que si nosotros somos quienes escribimos una funci贸n (declaramos) y a la vez los usuarios (llamamos a dicha funci贸n) no deberiamos forzar a la funci贸n a validar los tipos con los que se est谩 llamando, sino que deberiamos controlar que la estemos llamando correctamente. 

&nbsp;&nbsp;&nbsp;&nbsp; Obviamente no siempre es el caso. Si nosotros no somos los usuarios deberiamos hacer chequeos similares al implementeado en la funci贸n anterior. Por ejemplo, si lo que estamos desarrollando es una librer铆a; tendriamos que chequear que el usuario este llamando a la funci贸n de la API con los datos correctos para que no rompa m谩s abajo en el stack o mismo si estamos haciendo una API REST en donde los pedidos a un endpoint puede ser tan o m谩s creativos que nuestra imaginaci贸n.

&nbsp;&nbsp;&nbsp;&nbsp; Lo que **mypy** busca hacer es tratar de combinar los beneficios de tener un tipado est谩tico en el contexto de un lenguaje din谩mico. Para eso controla que los tipos que nosotros declaramos como hints coincidan con los de las variables con las que m谩s adelante se llaman. Por ejemplo miremos el archivo `classes.py` en donde declaramos dos clases equivalentes `A` y `B`, una de la manera cl谩sica de crear clases en Python y la otra con ayuda de la librer铆a estandar `dataclasses` (muy recomendada para evitar codear una y otra vez `__init__`).

````python
%%writefile classes.py

from dataclasses import dataclass


class A:
    def __init__(self, a: int, b: str, c: float):
        self.a = a
        self.b = b
        self.c = c


ia = A(a=1, b=1, c="asdad")


@dataclass
class B:
    a: int
    b: str
    c: float


ib = B(a=1, b=1, c="asdwqe")

print(f"ia: {ia}")
print(f"ib: {ib}")
````

## 1.4. Pydantic

&nbsp;&nbsp;&nbsp;&nbsp; Ahora bien. Como veniamos viendo, type hinting **NO** hace validaci贸n de tipos. Si queremos validar tipos tenemos que chequearlo durante runtime. Para esto podemos usar el viejo y querido `assert` como en el ejemplo anterior o bien, si estamos declarando tipos podemos hacerlo con `Pydantic`.

````python
!pip install pydantic
````

&nbsp;&nbsp;&nbsp;&nbsp; 驴C贸mo funciona? Nada mejor que un buen ejemplo autocontenido:

````python
from datetime import datetime
from typing import List, Optional
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name = 'Natalia Natalia'
    signup_ts: Optional[datetime] = None
    friends: List[int] = []

external_data = {'id': '123', 'signup_ts': '2019-06-01 12:22', 'friends': [1, '2', b'3', 4.9]}
user = User(**external_data)
print(user.id)
print(repr(user.signup_ts))
print(user.friends)
print(user.dict())
````

&nbsp;&nbsp;&nbsp;&nbsp; Como podemos ver, **pydantic** no solo valida que los tipos de los argumentos con los que estamos llamando sino que, en caso de que no sea v谩lido, trata de castearlo hac铆a el especificado en la declaraci贸n. Y digo trata porque no hace magia:

````python
from pydantic import ValidationError

external_data = {'id': 'Carlitos el Corrupto', 'signup_ts': 'ayer', 'friends': ["asd"]}
try:
    user = User(**external_data)
except ValidationError as e:
    print(e.json())
````

&nbsp;&nbsp;&nbsp;&nbsp; En este caso, no existe una forma sencilla o l贸gica de obtener un entero a partir de el string `"Carlitos el Corrupto"` por lo que falla en la validaci贸n al igual que en el resto de los campos. Esto es lo que pasa por ejemplo, si teniendo un endpoint de una API REST construida con [FastAPI](https://fastapi.tiangolo.com/) (que utiliza **pydantic** tras bambalinas) es llamado con un request sin el formato esperado.

&nbsp;&nbsp;&nbsp;&nbsp; Esto est谩 genial! 驴Por qu茅 no usarlo siempre? Bueno, por un lado, el concepto que hablamos anteriormente con respecto a cuando deberiamos validar data vs validar llamadas; y por otro, imperceptible al ojo, es menos performante que por ejemplo usar **dataclasses** (que no hace validaci贸n de tipos).


&nbsp;&nbsp;&nbsp;&nbsp; Algunos de los otros features interesantes de **pydantic** incluyen:
* Validators: Podemos validar, adem谩s del tipo, cualquier propiedad que se nos ocurran durante la instanciaci贸n de una clase.

````python
from pydantic import BaseModel, ValidationError, validator


class UserModel(BaseModel):
    name: str
    username: str

    @validator('name')
    def name_must_contain_carlitos(cls, v):
        if 'carlitos' not in v.lower():
            raise ValueError('only carlitos allowed!')
        return v.title()


user = UserModel(
    name='Carlitos el incorruptible',
    username='charlie123',
)
print(user)


try:
    UserModel(
        name='Otro Nombre',
        username='humai123',
    )
except ValidationError as e:
    print(e)
````

* Agregar restricciones a nivel de clase. Es decir para todos los argumentos.

````python
from pydantic import BaseModel, ValidationError


class Model(BaseModel):
    a: str
    b: str


    class Config:
        max_anystr_length = 10
        error_msg_templates = {
            'value_error.any_str.max_length': 'max_length:{limit_value}',
        }


try:
    Model(a='x' * 5, b='y' * 15)
except ValidationError as e:
    print(e)
````

* Creaci贸n de schemas en formato JSON a partir de los modelos.

````python
from pydantic import BaseModel, Field


class Gods(BaseModel):
    strengh_level: int
    size: float = None

class Carlitos(Gods):
    name: str = "The one and only"


class Universe(BaseModel):
    """
    This is the description of the main model
    """

    gods: Gods = Field(...)
    carlitos: Carlitos = Field(None, alias='Carlitos')
    snap: int = Field(
        42,
        title='The Snap',
        description='this is the value of snap',
        gt=30,
        lt=50,
    )

    class Config:
        title = 'The universe'


print(Universe.schema_json(indent=2))
````
